```python
import pandas as pd
import numpy as np
import streamlit as st
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import plotly.graph_objects as go
from datetime import datetime, timedelta

# -----------------------------
# 1 Generate Synthetic Dataset
# -----------------------------

def generate_dataset():

    np.random.seed(42)

    start_date = datetime(2024,1,1)
    days = 200

    dates = [start_date + timedelta(days=i) for i in range(days)]

    energy = []
    exam = []

    for i in range(days):

        base = 120

        noise = np.random.normal(0,8)

        # exam season boost
        if 70 < i < 90 or 150 < i < 170:
            exam_flag = 1
            value = base + 60 + noise
        else:
            exam_flag = 0
            value = base + noise

        energy.append(value)
        exam.append(exam_flag)

    df = pd.DataFrame({
        "date":dates,
        "energy_kwh":energy,
        "exam_period":exam
    })

    return df


df = generate_dataset()

df.set_index("date", inplace=True)

# -----------------------------
# 2 Forecast Model
# -----------------------------

model = ExponentialSmoothing(
    df["energy_kwh"],
    trend="add",
    seasonal=None
).fit()

forecast_days = 7

forecast = model.forecast(forecast_days)

predicted_energy = forecast.mean()

# -----------------------------
# 3 Streamlit Dashboard
# -----------------------------

st.title("📚 Library Energy Usage Forecast (Exam Period)")

st.write("Forecasting library electricity usage during exams using Exponential Smoothing.")

# Gauge Chart

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=predicted_energy,
    title={'text':"Predicted Avg Energy (kWh) Next Week"},
    gauge={
        'axis':{'range':[0,250]},
        'bar':{'color':"orange"},
        'steps':[
            {'range':[0,120],'color':"lightgreen"},
            {'range':[120,180],'color':"yellow"},
            {'range':[180,250],'color':"red"}
        ]
    }
))

st.plotly_chart(fig)

# -----------------------------
# 4 Historical Data Plot
# -----------------------------

st.subheader("Historical Energy Usage")

st.line_chart(df["energy_kwh"])

# -----------------------------
# 5 Forecast Plot
# -----------------------------

forecast_df = pd.DataFrame({
    "date":forecast.index,
    "forecast_energy":forecast.values
})

forecast_df.set_index("date", inplace=True)

st.subheader("Forecast for Next 7 Days")

st.line_chart(forecast_df)

# -----------------------------
# 6 Show Dataset
# -----------------------------

if st.checkbox("Show Dataset"):
    st.write(df)
```
